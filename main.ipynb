```python

# import data
import pandas as pd
df = pd.read_excel('uk_air_passengers.xlsx', sheet_name='in', header=0)

# check data
print(df)

# Drop null values
df = df.dropna()

# format date column
df['Date'] = pd.to_datetime(df['Date'], format='%d. %b %y')



# dataset reference
# https://www.kaggle.com/datasets/mohamedalishiha/uk-air-travel-passengers
```

```python

# 1. What is the average number of passengers per day?
# solution

avg_passengers_per_day = df['Passengers'].mean()

print('The average number of passengers per day is', avg_passengers_per_day)
```

```python
# 2. Visualize the number of passengers per year
# solution

import matplotlib.pyplot as plt
plt.plot(df['Date'], df['Passengers'])
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers per year')
plt.show()
```

```python
# Question 3. Group the data by year and month and visualize the number of passengers per month
# solution

# group by year and month
df['year_month'] = df['year'].astype(str) + '-' + df['month'].astype(str)
df_grouped = df.groupby('year_month').sum()

# plot
plt.plot(df_grouped.index, df_grouped['Passengers'])
plt.xlabel('Year-Month')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers per month')
plt.show()
```

```python
# Question 4. In a pie chart, visualize, compare the number of passengers of each year

# solution

# group by year
df_grouped = df.groupby('year').sum()

# plot
plt.pie(df_grouped['Passengers'], labels=df_grouped.index, autopct='%1.1f%%')
plt.title('Number of Passengers per year')
plt.show()
```

```python
# Question 5. For each month, for each year, show mean, median, mode, standard deviation, variance, min, max, 25%, 50%, 75% of the number of passengers

# solution

# group by year and month
df['year_month'] = df['year'].astype(str) + '-' + df['month'].astype(str)
df_grouped = df.groupby('year_month').describe()


# print
print(df_grouped['Passengers'])
```

```python
# Draw line chart of increasing passengers yearly
# solution

# group by year
df_grouped = df.groupby('year').sum()

# plot
plt.plot(df_grouped.index, df_grouped['Passengers'])
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers per year')
plt.show()
```
